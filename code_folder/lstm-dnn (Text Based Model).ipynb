```python

word_vectors_for_training = pd.np.insert(
    model.wv.vectors,   
    0, 
    pd.np.random.uniform(model.wv.vectors.min(),model.wv.vectors.max(),100),
    axis=0
)

word_vectors_for_training = pd.np.insert(
    word_vectors_for_training,   
    0, 
    pd.np.zeros(100),
    axis=0
)
word_vectors_for_training = torch.FloatTensor(word_vectors_for_training)
word_vectors_for_training.shape
```

```python
class FirstNeuralNetwork(nn.Module): 
    def __init__(self, input_size, hidden_size, output_size):
        super().__init__()
        
        self.embedding_layer = nn.EmbeddingBag.from_pretrained(word_vectors_for_training, mode="mean") 
        self.layer1 = nn.Linear(input_size, hidden_size) 
        self.relu1 = nn.ReLU() 
        self.layer2 = nn.Linear(hidden_size, hidden_size//4) #out features : 64 
        self.relu2 = nn.ReLU() 
        self.layer3 = nn.Linear(hidden_size//4, hidden_size // 8) #out features : 32 
        self.relu3 = nn.ReLU() 
        self.layer4 = nn.Linear(hidden_size//8, output_size)
        self.relu4 = nn.ReLU()

    def forward(self, data_input):
        embedded_data_input = self.embedding_layer(data_input)
        output_layer1 = self.layer1(embedded_data_input)
        output_layer1 = self.relu1(output_layer1)
        output_layer2 = self.layer2(output_layer1)
        output_layer2 = self.relu2(output_layer2)
        output_layer3 = self.layer3(output_layer2)
        output_layer3 = self.relu3(output_layer3)
        output_layer4 = self.layer4(output_layer3)
        output_layer4 = self.relu4(output_layer4)
        return output_layer4.squeeze()
```

```python
# finalize the training data
#training_sentence_data = torch.stack(training_sentence_data)

# Define the network
first_neural_network =FirstNeuralNetwork(input_size = 100, hidden_size= 256, output_size =1).cuda()
```

```python
import torch.utils.data as data
```

```python
dataset = data.TensorDataset(training_sentence_data, torch.FloatTensor(df.retweet_count.values))
dataloader = data.DataLoader(dataset, batch_size=256)
```

```python
import torch.optim as optim
optimizer = optim.Adam(first_neural_network.parameters(), lr=0.001)
loss_fn = torch.nn.L1Loss()
```

```python
from math import nan, isnan
def py_sum(alist):
    
    output = 0.0
    alist_filtered = [x for x in alist if isnan(x) == False]
    for i in alist_filtered:
        
        
        output += i
        #print("out : ", output, " when i is : ", i)
    #print("output : ", output)
    return output
```

```python
#progress bar
import tqdm

# number of epochs
for n in tqdm.tqdm(range(100)):
    avg_loss = []
    counter = 0
    for batch in dataloader:
        # batch [0] - the sentence data or X
        # batch [1] - the label for each sentence or Y
        
        # for every back pass you need to zero out the optimizer
        # less you get residual gradients
        optimizer.zero_grad()
        
        # pass the model into the batch
        # this line is the same as calling first_neural_network.foward(batch)
        # yay shortcuts
        output = first_neural_network(batch[0].cuda())
        counter += 1
        
        # Calculate the loss function
        #loss = torch.nn.functional.mse_loss(output, batch[1])
        loss = loss_fn(output, batch[1].cuda())
        
        # Save the loss for each epoch
        avg_loss.append(loss.item())
        
        # Tell pytorch to calculate the gradient
        loss.backward()
        
        # tell pytorch to pass the gradients back into the model
        optimizer.step()
    
   
    #loss_sum = py_sum(avg_loss)
    #length = len(avg_loss)
   
    #print(loss_sum / length)
    alist_filtered = [x for x in avg_loss if isnan(x) == False]
    #print(pd.np.mean(alist_filtered))
    
    #accuracy
    correct = (output == batch[1].cuda()).float().sum()
    print("Epoch {}/{}, Loss: {:.3f}, Accuracy: {:.3f}".format(n+1,100, pd.np.mean(alist_filtered), correct/counter))
```
